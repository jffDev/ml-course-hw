1.	Методами машинного обучения (не статистическими тестами) показать, что разбиение на трейн и тест репрезентативно.

Можно рассмотреть несколько подходов, чтобы убедиться, что train и test репрезентативны. Например, можно использовать методы кластеризации, обучить классификатор или применить более экзотические методы с понижением размерности.
Например, если берем метод кластеризации, то смотрим на кластеры train и test и ожидаем, что они будут примерно одинаковы. Следовательно, делаем вывод, что наше разбиение репрезентативно и нет в противном случае.
Если же воспользуемся методом обучения классификатора, то в случае репрезентативного разбиения классификатор не сможет отличать данные между train и test. Если же мы применим какой-нибудь рандомный классификатор с зафиксированным уровнем точности, и наш классификатор вдруг на таком распределении покажет результат выше, то тогда можно считать, что данные в train и test не репрезентативны.

2.	Есть кластеризованный датасет на 4 кластера (1, 2, 3, 4). Бизнес аналитики посчитали, что самым прибыльным является кластер 2. Каждый клиент представлен в виде 10ти-мертного вектора, где первые 6 значений транзакции, а оставшиеся: возраст, пол, социальный статус (женат (замужем)/неженат (не замужем)), количество детей. Нужно поставить задачу оптимизации для каждого клиента не из кластера 2 так, чтобы увидеть, как должен начать вести себя клиент, чтобы перейти в кластер 2.

Наиболее логичным видится поработать сначала со вторым кластером, т.е. самым прибыльным. Например, мы точно знаем, что некоторые параметры не относятся к транзакциям, а именно последние 4.
Поэтому можно попробовать дополнительно кластеризовать наш второй кластер по последним 4 признакам, т.е. возраст, пол, социальный статус, количество детей. После этого мы можем для клиентов из кластеров 1,3 и 4 определить их "расстояние" до каждого из сформированных кластеров кластера 2 и уже по полученным данным, т.е. по самому близкому "расстоянию" определить пул транзакций для клиентов из неприбыльных кластеров. После чего рекомендательными алгоритмами (например, по алгоритму коллаборативной фильтрации) определить, какие товары порекомендовать клиенту, тем самым потенциально повышая количество его транзакций и возможный переход во второй кластер.

3.	Что лучше 2 модели случайного леса по 500 деревьев или одна на 1000, при условии, что ВСЕ параметры кроме количества деревьев одинаковы?

Так как по условию даны ограничения, что все параметры равны, кроме количества деревьев, то несложно исследовать этот вопрос программным путем. Но если же рассуждать теоретически и посмотреть реализацию RandomForestClassifier пакета sklearn, то можно сразу определить, что в случае двух моделей случайного леса по 500 деревьев мы получим по факту идентичные модели с одинаковыми предсказаниями. Это напрямую связано с параметром random_state.
Если посмотрим на определение данного параметра в документации, а именно "Управляет как случайностью начальной загрузки выборок, используемых при построении деревьев, так и выборкой объектов, которые необходимо учитывать при поиске наилучшего разделения на каждом узле", то несложно сделать вывод, что в двух моделях будет очень схожая или идентичная группировка признаков.

Т.е. по факту у нас становится сравнение между моделью случайного леса на 500 деревьев и на 1000 деревьев.
Можно в целом сказать, что на 1000 деревьев даст более точную и стабильную оценку, следовательно, одна модель на 1000 деревьев более предпочтительна.

4.	В наличии датасет с данными по дефолту клиентов. Как, имея в инструментарии только алгоритм kmeans получить вероятность дефолта нового клиента.

Итак, у нас дан датасет с данными дефолта клитентов. Можно использовать только алгоритм kmeans. Исходя из этих ограничений можно пойти путем применения алгоритма kmeans к нашему датасету. Далее по каждому кластеру считаем вероятность дефолта по простому алгоритму: количество дефолтов к общему количеству клиентов в этом кластере. Далее алгоритм действий может быть такой. Приходит новый клиент, мы определяем принадлежность этого нового клиента к имеющимся кластерам и берем ту вероятность дефолта, которая была у кластера. После чего присваиваем клиенту это значение вероятности дефолта.

5.	Есть выборка клиентов с заявкой на кредитный продукт. Датасет состоит из персональных данных: возраст, пол и т.д. Необходимо предсказывать доход клиента, который представляет собой непрерывные данные, но сделать это нужно используя только модель классификации.

Первоначально необходимо определить бизнес-критерии по диапазонам дохода, что дает нам классы для решения задачи классификации.
Далее решение поставленной задачи сводится к решению задачи классификации:
1. Мы получили необходимые диапазоны и сформировали по ним классы.
2. Выбираем или реализуем модель классификатора. Тут выбор может быть довольно большой, зависит от дополнительных требований, например, интерпретируемости результата.
3. Возможно датасет необходимо будет предварительно обработать(провести анализ данных датасета и подготовить для использования в выбранной модели).
4. Обучаем модель классификации, после чего наша модель может предсказывать вероятность того или иного класса.
5. После чего мы можем определить среднее значение дохода для каждого нового клиента как математическое ожидание непрерывной случайной величины.
